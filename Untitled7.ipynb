```python
from statsmodels.tsa.seasonal import seasonal_decompose as sd
sd_SF = sd(df['TG'], freq=365)
sd_M = sd(df['TN'], freq=365)

_=plt.figure(figsize=(15,10))
ax1=plt.subplot(311)
_=ax1.plot(sd_SF.trend, label='TG', alpha=0.7)
_=ax1.plot(sd_M.trend, label='TN', alpha=0.7)
_=plt.legend()
ax2=plt.subplot(312)
_=ax2.plot(sd_SF.seasonal, label='TG', alpha=0.7)
_=ax2.plot(sd_M.seasonal, label='TN', alpha=0.7)
_=plt.legend()
ax3=plt.subplot(313)
_=ax3.plot(sd_SF.resid, label='TG', alpha=0.7)
_=ax3.plot(sd_M.resid, label='TN', alpha=0.7)
_=plt.legend()
```
